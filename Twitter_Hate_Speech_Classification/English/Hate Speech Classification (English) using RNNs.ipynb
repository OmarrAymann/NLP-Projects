# `04` **Cleaning and Preprocessing**

### Preprocessing

```python
def clean_text(tweet):
    tweet = str(tweet).lower()
    tweet = re.sub(r'http\S+', '',tweet)
    tweet = re.sub(r"\d+", "",tweet)
    tweet = re.sub(r'@\S+', '',tweet)
    tweet = re.sub(r'#(\S+)', r'\1',tweet)
    tweet = re.sub(r'\s+', ' ',tweet)
    tweet = re.sub(r'[^a-zA-Z0-9\s]', '', tweet)
    tokens = word_tokenize(tweet)
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    lemmatizer = WordNetLemmatizer()
    lemmatizer_word =  [lemmatizer.lemmatize(word) for word in filtered_tokens]
    cleaned_text = ' '.join(lemmatizer_word)
    return cleaned_text
```

```python
X_train = X_train.apply(clean_text)
X_test = X_test.apply(clean_text)

X_train.head(10)
```

## `+` Set Global Parameters

```python
BATCH_SIZE  = 64
EPOCHS      = 10
early_stop = EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True)
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)

word_size = len(tokenizer.word_index) + 1

X_train_pad = tokenizer.texts_to_sequences(X_train)
X_test_pad = tokenizer.texts_to_sequences(X_test)

max_len = 150
X_train_pad = pad_sequences(X_train_pad, maxlen=max_len)
X_test_pad = pad_sequences(X_test_pad, maxlen=max_len)
```

## `i` **Model Design**
- **CNN**

```python
model = Sequential([
    Embedding(input_dim=word_size, output_dim=100),
    Conv1D(filters=64, kernel_size=3, activation='relu'),
    GlobalMaxPooling1D(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.build(input_shape=(None, max_len))
model.summary()
```

## `ii` Model Compilation & Training

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
history = model.fit(X_train_pad, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, validation_split=0.2, callbacks=[early_stop], verbose=1)
```

## `iii` Model Evaluation

```python
loss, accuracy = model.evaluate(X_test_pad, y_test)
print(f"Loss: {loss}")
print(f"Accuracy: {accuracy}")
```

## `iv` Result Visualization

```python
plt.plot(history.history['accuracy'], label='Training')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['loss'], label='Training')
plt.plot(history.history['val_loss'], label='Validation')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

```python
y_pred = model.predict(X_test_pad)
y_pred = (y_pred > 0.5).astype(int)
print(classification_report(y_test, y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='magma', xticklabels=['Not Hate Speech', ' Hate Speech'], yticklabels=['Not Hate Speech', ' Hate Speech'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
def test_tweet(tweet):
    cleaned_tweet = clean_text(tweet)  
    sequence = tokenizer.texts_to_sequences([cleaned_tweet])
    padded_sequence = pad_sequences(sequence, maxlen=150, padding='post', truncating='post')
    prediction = model.predict(padded_sequence)[0][0]
    if prediction > 0.5: 
        pred = "Hate tweet" 
    else: 
        pred = "No Hate tweet"

    return {
        "cleaned_text": cleaned_tweet,
        "probability": float(prediction),
        "prediction": pred
    }
```

```python
input = "leo messi is the best player in the world #messi #GOAT #football" 
result = test_tweet(input)
print(f"Tweet: {input}\n")
print(f"Cleaned tweet: {result['cleaned_text']}\n")
print(f"Prediction: {result['prediction']}\n")
```
